```python
from rdkit import Chem
from rdkit.Chem import AllChem, MACCSkeys

# 예시로 3개의 SMILES 표기법 화합물을 사용합니다.
smiles_list = ["c1ccccc1", "CCO", "CC(=O)O"]

# 각 화합물에 대해 결과를 저장할 빈 리스트를 생성합니다.
maccs_keys_results = []
ecfp_results = []
fcfp_results = []

radius = 2
n_bits = 1024

# 각 화합물에 대해 MACCS Keys, ECFP, FCFP를 계산합니다.
for smiles in smiles_list:
    mol = Chem.MolFromSmiles(smiles)
    
    maccs_keys = MACCSkeys.GenMACCSKeys(mol)
    maccs_keys_list = list(maccs_keys)
    maccs_keys_results.append(maccs_keys_list)
    
    ecfp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=n_bits)
    ecfp_list = list(ecfp)
    ecfp_results.append(ecfp_list)
    
    fcfp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=n_bits, useFeatures=True)
    fcfp_list = list(fcfp)
    fcfp_results.append(fcfp_list)

# 각 화합물에 대한 결과를 출력합니다.
for i, smiles in enumerate(smiles_list):
    print(f"Results for compound {i+1} (SMILES: {smiles}):")
    print("MACCS Keys Fingerprint: ", maccs_keys_results[i])
    print("ECFP: ", ecfp_results[i])
    print("FCFP: ", fcfp_results[i])
    print()
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem, MACCSkeys, DataStructs

smiles_list = ["c1ccccc1", "CCO", "CC(=O)O"]

# 각 화합물에 대해 결과를 저장할 빈 리스트를 생성합니다.
ecfp_results = []

radius = 2
n_bits = 1024

# 각 화합물에 대해 ECFP를 계산합니다.
for smiles in smiles_list:
    mol = Chem.MolFromSmiles(smiles)
    ecfp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=n_bits)
    ecfp_results.append(ecfp)

# 화합물 간의 Tanimoto 유사성 계산
similarity_matrix = []
for i in range(len(ecfp_results)):
    row = []
    for j in range(len(ecfp_results)):
        similarity = DataStructs.TanimotoSimilarity(ecfp_results[i], ecfp_results[j])
        row.append(similarity)
    similarity_matrix.append(row)

# 화합물 간의 유사성 출력
print("Tanimoto Similarity Matrix:")
for row in similarity_matrix:
    print(row)
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem, DataStructs
import numpy as np

def dice_similarity(fp1, fp2):
    intersection = np.bitwise_and(fp1, fp2)
    return 2 * np.sum(intersection) / (np.sum(fp1) + np.sum(fp2))

def cosine_similarity(fp1, fp2):
    dot_product = np.dot(fp1, fp2)
    magnitude = np.linalg.norm(fp1) * np.linalg.norm(fp2)
    return dot_product / magnitude

def soergel_distance(fp1, fp2):
    intersection = np.bitwise_and(fp1, fp2)
    union = np.bitwise_or(fp1, fp2)
    return 1 - np.sum(intersection) / np.sum(union)

smiles_list = ["c1ccccc1", "CCO", "CC(=O)O"]
ecfp_results = []

radius = 2
n_bits = 1024

for smiles in smiles_list:
    mol = Chem.MolFromSmiles(smiles)
    ecfp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=n_bits)
    ecfp_results.append(ecfp)

# 화합물 간의 Dice 유사성 계산
dice_matrix = []
for i in range(len(ecfp_results)):
    row = []
    for j in range(len(ecfp_results)):
        similarity = dice_similarity(ecfp_results[i], ecfp_results[j])
        row.append(similarity)
    dice_matrix.append(row)

# 화합물 간의 Cosine 유사성 계산
cosine_matrix = []
for i in range(len(ecfp_results)):
    row = []
    for j in range(len(ecfp_results)):
        similarity = cosine_similarity(ecfp_results[i], ecfp_results[j])
        row.append(similarity)
    cosine_matrix.append(row)

# 화합물 간의 Soergel 거리 계산
soergel_matrix = []
for i in range(len(ecfp_results)):
    row = []
    for j in range(len(ecfp_results)):
        distance = soergel_distance(ecfp_results[i], ecfp_results[j])
        row.append(distance)
    soergel_matrix.append(row)

# 화합물 간의 유사성/거리 출력
print("Dice Similarity Matrix:")
for row in dice_matrix:
    print(row)
print("\nCosine Similarity Matrix:")
for row in cosine_matrix:
    print(row)
print("\nSoergel Distance Matrix:")
for row in soergel_matrix:
    print(row)
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from sklearn.cluster import KMeans
import numpy as np

# 예시로 10개의 SMILES 표기법 화합물을 사용합니다.
smiles_list = [
    "c1ccccc1",
    "CCO",
    "CC(=O)O",
    "C1=CC=C(C=C1)O",
    "C1=CC(=CC=C1)O",
    "CCCC",
    "CCN",
    "C1=CC(=CC=C1)F",
    "CCOC(=O)C",
    "CC(=O)N"
]

ecfp_results = []

radius = 2
n_bits = 1024

for smiles in smiles_list:
    mol = Chem.MolFromSmiles(smiles)
    ecfp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=n_bits)
    ecfp_results.append(ecfp)

# fingerprint 비트 벡터를 numpy 배열로 변환합니다.
X = np.array(ecfp_results)

# 클러스터 수 설정
n_clusters = 3

# k-means 알고리즘으로 클러스터링 수행
kmeans = KMeans(n_clusters=n_clusters, random_state=42).fit(X)

# 각 화합물에 대한 클러스터 레이블을 출력합니다.
for i, label in enumerate(kmeans.labels_):
    print(f"Compound {i+1} (SMILES: {smiles_list[i]}) is in cluster {label+1}")
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

smiles_list = [
    "c1ccccc1",
    "CCO",
    "CC(=O)O",
    "C1=CC=C(C=C1)O",
    "C1=CC(=CC=C1)O",
    "CCCC",
    "CCN",
    "C1=CC(=CC=C1)F",
    "CCOC(=O)C",
    "CC(=O)N"
]

ecfp_results = []

radius = 2
n_bits = 1024

for smiles in smiles_list:
    mol = Chem.MolFromSmiles(smiles)
    ecfp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=n_bits)
    ecfp_results.append(ecfp)

X = np.array(ecfp_results)

# 클러스터 수 범위 설정
min_clusters = 1
max_clusters = len(smiles_list)

# 클러스터링을 수행하고 각 클러스터 수에 대한 WCSS를 계산합니다.
wcss = []
for i in range(min_clusters, max_clusters+1):
    kmeans = KMeans(n_clusters=i, random_state=42).fit(X)
    wcss.append(kmeans.inertia_)

# WCSS 그래프를 그립니다.
plt.plot(range(min_clusters, max_clusters+1), wcss, marker='o')
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.DataStructs import cDataStructs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# 예제용 화합물 SMILES
smiles_list = [
    "CCO",
    "CCCO",
    "CCCCO",
    "CC(N)CO",
    "CCNCCC",
    "CCCCN",
    "CCCCCO",
    "CC(C)CO",
    "CCCC(C)O",
    "CC(C)C(O)"
]

# 화합물을 ECFP4 형태로 변환
molecules = [Chem.MolFromSmiles(smiles) for smiles in smiles_list]
fps = [AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=2048) for mol in molecules]

# 벡터 형태로 변환
X = []
for fp in fps:
    arr = np.zeros((1,))
    cDataStructs.ConvertToNumpyArray(fp, arr)
    X.append(arr)
X = np.array(X)

# Elbow Method를 사용하여 적절한 클러스터 수 찾기
distortions = []
K = range(1, len(smiles_list))
for k in K:
    kmeanModel = KMeans(n_clusters=k, random_state=42)
    kmeanModel.fit(X)
    distortions.append(kmeanModel.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(K, distortions, 'bx-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Distortion')
plt.title('Elbow Method showing the optimal k')
plt.show()

# 적정한 클러스터 수 결정
optimal_k = distortions.index(min(distortions)) + 1

# K-means 클러스터링 수행
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(X)

# 결과 출력
print(f"Optimal number of clusters: {optimal_k}")
print("Cluster assignments:")
for i, label in enumerate(kmeans.labels_):
    print(f"Compound {i+1}: {smiles_list[i]} -> Cluster {label+1}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.DataStructs import cDataStructs
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

# 예제용 화합물 SMILES
smiles_list = [
    "CCO",
    "CCCO",
    "CCCCO",
    "CC(N)CO",
    "CCNCCC",
    "CCCCN",
    "CCCCCO",
    "CC(C)CO",
    "CCCC(C)O",
    "CC(C)C(O)"
]

# 화합물을 ECFP4 형태로 변환
molecules = [Chem.MolFromSmiles(smiles) for smiles in smiles_list]
fps = [AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=1048) for mol in molecules]

# 벡터 형태로 변환
X = []
for fp in fps:
    arr = np.zeros((1,))
    cDataStructs.ConvertToNumpyArray(fp, arr)
    X.append(arr)
X = np.array(X)

# K-means 모델
model = KMeans(random_state=42)

# 데이터 샘플 수보다 작은 클러스터 수 범위 설정
max_k = min(len(X), 101)

# Elbow Method를 사용하여 적절한 클러스터 수 찾기
visualizer = KElbowVisualizer(model, k=(1, max_k), timings=False)
visualizer.fit(X)
visualizer.show()

# 적절한 클러스터 수 결정
optimal_k = visualizer.elbow_value_

# K-means 클러스터링 수행
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(X)

# 결과 출력
print(f"Optimal number of clusters: {optimal_k}")
print("Cluster assignments:")
for i, label in enumerate(kmeans.labels_):
    print(f"Compound {i+1}: {smiles_list[i]} -> Cluster {label+1}")
```

```python
from rdkit.Chem import AllChem
from rdkit import DataStructs
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Descriptors
from rdkit.Chem import PandasTools
from rdkit.Chem import rdMolDescriptors

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from yellowbrick.cluster import KElbowVisualizer

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 예제용 화합물 SMILES 문자열
smiles_list = [
    "CCO",
    "CCC",
    "CCN",
    "C(=O)O",
    "CC#N",
    "CCCl",
    "CCBr",
    "CC(F)(F)F",
    "C#CC",
    "C=C",
    "C1CCCCC1",
    "c1ccccc1",
    "CCc1ccccc1",
    "CCOCC",
    "CCN(CC)CC",
    "c1ccncc1",
    "C1CCOCC1",
    "C1CCNCC1",
]

# SMILES 문자열을 RDKit 분자로 변환
mols = [Chem.MolFromSmiles(smiles) for smiles in smiles_list]

# 분자별 ECFP4 fingerprints 계산
fp_list = [AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=1024) for mol in mols]

# fingerprints를 NumPy 배열로 변환
X = np.array([np.array(fp) for fp in fp_list], dtype=np.float32)

# Elbow Method를 사용하여 적절한 클러스터 수 찾기
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1, len(X) // 2 + 1), timings=False)
visualizer.fit(X)
visualizer.show()

# 적절한 클러스터 수 결정
optimal_k = visualizer.elbow_value_

# KMeans 클러스터링 수행
kmeans = KMeans(n_clusters=optimal_k, random_state=42).fit(X)

# 클러스터 라벨 할당
labels = kmeans.labels_

# 결과 출력
for i, (smiles, label) in enumerate(zip(smiles_list, labels)):
    print(f"화합물 {i + 1} ({smiles}): 클러스터 {label}")
```
