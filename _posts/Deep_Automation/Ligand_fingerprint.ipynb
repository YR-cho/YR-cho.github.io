```python
from rdkit import Chem
from rdkit.Chem import AllChem, MACCSkeys, DataStructs

smiles_list = ["c1ccccc1", "CCO", "CC(=O)O"]

# 각 화합물에 대해 결과를 저장할 빈 리스트를 생성합니다.
ecfp_results = []

radius = 2
n_bits = 1024

# 각 화합물에 대해 ECFP를 계산합니다.
for smiles in smiles_list:
    mol = Chem.MolFromSmiles(smiles)
    ecfp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=n_bits)
    ecfp_results.append(ecfp)

# 화합물 간의 Tanimoto 유사성 계산
similarity_matrix = []
for i in range(len(ecfp_results)):
    row = []
    for j in range(len(ecfp_results)):
        similarity = DataStructs.TanimotoSimilarity(ecfp_results[i], ecfp_results[j])
        row.append(similarity)
    similarity_matrix.append(row)

# 화합물 간의 유사성 출력
print("Tanimoto Similarity Matrix:")
for row in similarity_matrix:
    print(row)
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem, DataStructs
import numpy as np

def dice_similarity(fp1, fp2):
    intersection = np.bitwise_and(fp1, fp2)
    return 2 * np.sum(intersection) / (np.sum(fp1) + np.sum(fp2))

def cosine_similarity(fp1, fp2):
    dot_product = np.dot(fp1, fp2)
    magnitude = np.linalg.norm(fp1) * np.linalg.norm(fp2)
    return dot_product / magnitude

def soergel_distance(fp1, fp2):
    intersection = np.bitwise_and(fp1, fp2)
    union = np.bitwise_or(fp1, fp2)
    return 1 - np.sum(intersection) / np.sum(union)

smiles_list = ["c1ccccc1", "CCO", "CC(=O)O"]
ecfp_results = []

radius = 2
n_bits = 1024

for smiles in smiles_list:
    mol = Chem.MolFromSmiles(smiles)
    ecfp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=n_bits)
    ecfp_results.append(ecfp)

# 화합물 간의 Dice 유사성 계산
dice_matrix = []
for i in range(len(ecfp_results)):
    row = []
    for j in range(len(ecfp_results)):
        similarity = dice_similarity(ecfp_results[i], ecfp_results[j])
        row.append(similarity)
    dice_matrix.append(row)

# 화합물 간의 Cosine 유사성 계산
cosine_matrix = []
for i in range(len(ecfp_results)):
    row = []
    for j in range(len(ecfp_results)):
        similarity = cosine_similarity(ecfp_results[i], ecfp_results[j])
        row.append(similarity)
    cosine_matrix.append(row)

# 화합물 간의 Soergel 거리 계산
soergel_matrix = []
for i in range(len(ecfp_results)):
    row = []
    for j in range(len(ecfp_results)):
        distance = soergel_distance(ecfp_results[i], ecfp_results[j])
        row.append(distance)
    soergel_matrix.append(row)

# 화합물 간의 유사성/거리 출력
print("Dice Similarity Matrix:")
for row in dice_matrix:
    print(row)
print("\nCosine Similarity Matrix:")
for row in cosine_matrix:
    print(row)
print("\nSoergel Distance Matrix:")
for row in soergel_matrix:
    print(row)
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

smiles_list = [
    "c1ccccc1",
    "CCO",
    "CC(=O)O",
    "C1=CC=C(C=C1)O",
    "C1=CC(=CC=C1)O",
    "CCCC",
    "CCN",
    "C1=CC(=CC=C1)F",
    "CCOC(=O)C",
    "CC(=O)N"
]

ecfp_results = []

radius = 2
n_bits = 1024

for smiles in smiles_list:
    mol = Chem.MolFromSmiles(smiles)
    ecfp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=n_bits)
    ecfp_results.append(ecfp)

X = np.array(ecfp_results)

# 클러스터 수 범위 설정
min_clusters = 1
max_clusters = len(smiles_list)

# 클러스터링을 수행하고 각 클러스터 수에 대한 WCSS를 계산합니다.
wcss = []
for i in range(min_clusters, max_clusters+1):
    kmeans = KMeans(n_clusters=i, random_state=42).fit(X)
    wcss.append(kmeans.inertia_)

# WCSS 그래프를 그립니다.
plt.plot(range(min_clusters, max_clusters+1), wcss, marker='o')
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.DataStructs import cDataStructs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# 예제용 화합물 SMILES
smiles_list = [
    "CCO",
    "CCCO",
    "CCCCO",
    "CC(N)CO",
    "CCNCCC",
    "CCCCN",
    "CCCCCO",
    "CC(C)CO",
    "CCCC(C)O",
    "CC(C)C(O)"
]

# 화합물을 ECFP4 형태로 변환
molecules = [Chem.MolFromSmiles(smiles) for smiles in smiles_list]
fps = [AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=2048) for mol in molecules]

# 벡터 형태로 변환
X = []
for fp in fps:
    arr = np.zeros((1,))
    cDataStructs.ConvertToNumpyArray(fp, arr)
    X.append(arr)
X = np.array(X)

# Elbow Method를 사용하여 적절한 클러스터 수 찾기
distortions = []
K = range(1, len(smiles_list))
for k in K:
    kmeanModel = KMeans(n_clusters=k, random_state=42)
    kmeanModel.fit(X)
    distortions.append(kmeanModel.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(K, distortions, 'bx-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Distortion')
plt.title('Elbow Method showing the optimal k')
plt.show()

# 적정한 클러스터 수 결정
optimal_k = distortions.index(min(distortions)) + 1

# K-means 클러스터링 수행
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(X)

# 결과 출력
print(f"Optimal number of clusters: {optimal_k}")
print("Cluster assignments:")
for i, label in enumerate(kmeans.labels_):
    print(f"Compound {i+1}: {smiles_list[i]} -> Cluster {label+1}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.DataStructs import cDataStructs
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

# 예제용 화합물 SMILES
smiles_list = [
    "CCO",
    "CCCO",
    "CCCCO",
    "CC(N)CO",
    "CCNCCC",
    "CCCCN",
    "CCCCCO",
    "CC(C)CO",
    "CCCC(C)O",
    "CC(C)C(O)"
]

# 화합물을 ECFP4 형태로 변환
molecules = [Chem.MolFromSmiles(smiles) for smiles in smiles_list]
fps = [AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=1048) for mol in molecules]

# 벡터 형태로 변환
X = []
for fp in fps:
    arr = np.zeros((1,))
    cDataStructs.ConvertToNumpyArray(fp, arr)
    X.append(arr)
X = np.array(X)

# K-means 모델
model = KMeans(random_state=42)

# 데이터 샘플 수보다 작은 클러스터 수 범위 설정
max_k = min(len(X), 101)

# Elbow Method를 사용하여 적절한 클러스터 수 찾기
visualizer = KElbowVisualizer(model, k=(1, max_k), timings=False)
visualizer.fit(X)
visualizer.show()

# 적절한 클러스터 수 결정
optimal_k = visualizer.elbow_value_

# K-means 클러스터링 수행
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(X)

# 결과 출력
print(f"Optimal number of clusters: {optimal_k}")
print("Cluster assignments:")
for i, label in enumerate(kmeans.labels_):
    print(f"Compound {i+1}: {smiles_list[i]} -> Cluster {label+1}")
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from scipy.spatial import Voronoi, voronoi_plot_2d
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 랜덤 화합물 데이터 생성
np.random.seed(42)
data = np.random.random((50, 4))

# 적절한 클러스터 수 찾기 (Elbow Method)
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# K-means 클러스터링 수행
n_clusters = 3 # 적절한 클러스터 수를 선택
kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42)
clusters = kmeans.fit_predict(data)

# PCA를 사용한 차원 축소
pca = PCA(n_components=2)
pca_data = pca.fit_transform(data)

# 2D Voronoi plot
vor = Voronoi(pca_data)
fig, ax = plt.subplots()
voronoi_plot_2d(vor, ax=ax)
ax.scatter(pca_data[:, 0], pca_data[:, 1], c=clusters, cmap='viridis')
ax.set_title('Voronoi Plot')
plt.show()

# 등고선 플롯
fig = plt.figure()
plt.contourf(pca_data[:, 0].reshape(-1, 1), pca_data[:, 1].reshape(-1, 1), clusters.reshape(-1, 1), cmap='viridis')
plt.colorbar()
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=clusters, cmap='viridis')
plt.title('Contour Plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from scipy.spatial import Voronoi, voronoi_plot_2d, ConvexHull

# 랜덤하게 화합물 50개 생성
X, _ = make_blobs(n_samples=50, centers=3, random_state=42)

# Elbow 방법으로 적절한 클러스터 수 찾기
distortions = []
K = range(1, 11)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(X)
    distortions.append(kmeanModel.inertia_)

# Elbow point를 찾아 클러스터 개수로 설정
# Elbow point를 찾아 클러스터 개수로 설정
elbow_point = max(3, np.argmax(np.diff(distortions, 2)) + 2)
kmeans = KMeans(n_clusters=elbow_point)
clusters = kmeans.fit_predict(X)

# PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(X)

# Voronoi diagram 그리기
vor = Voronoi(kmeans.cluster_centers_)
fig, ax = plt.subplots()
voronoi_plot_2d(vor, ax=ax)
ax.scatter(pca_data[:, 0], pca_data[:, 1], c=clusters, cmap='viridis')

# Voronoi regions의 면적 계산
areas = []
for region in vor.regions:
    if not -1 in region and len(region) > 0:
        polygon = [vor.vertices[i] for i in region]
        hull = ConvexHull(polygon)
        area = hull.volume
        areas.append(area)

print("각 클러스터의 면적:", areas)
plt.show()
```
